```python
import pandas as pd
import matplotlib.pyplot as plt
url ="/kaggle/student_data.csv"
da = pd.read_csv(url)
print(da)
```

```python
da['age']
```

```python
plt.figure(figsize=(12, 4))

x = range(da['age'].size)
y = da['age']

plt.scatter(x, y, marker='x', s=1)
plt.title('Ages of student')
plt.xlabel('ID of student')
plt.ylabel('age')

plt.show()
```
